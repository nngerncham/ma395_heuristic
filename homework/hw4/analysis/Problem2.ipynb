```python
import seaborn as sns
import matplotlib.pylab as plt
import pandas as pd
```

```python
def read_results(path):
    data = pd.read_csv(path).set_index("Generation")
    data["Optimal"] = 66000
    return data
```

```python
sns.lineplot(read_results("../results/p2/part4.csv"), errorbar="sd")
plt.ylim(60000, 70000)
plt.savefig("../images/p2/part4.png", bbox_inches="tight")
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
probs = [[0.03, 0.05], [0.075, 0.1]]
for i in range(2):
    for j in range(2):
        axes[i, j].set_title(f"Mutation probability: {probs[i][j]}")
        if i == 0:
            axes[i, j].set_ylim(50000, 68000)
        else:
            axes[i, j].set_ylim(60000, 68000)
sns.lineplot(read_results("../results/p2/part5pr0.03.csv"), errorbar="sd", ax=axes[0, 0])
sns.lineplot(read_results("../results/p2/part5pr0.05.csv"), errorbar="sd", ax=axes[0, 1])
sns.lineplot(read_results("../results/p2/part5pr0.075.csv"), errorbar="sd", ax=axes[1, 0])
sns.lineplot(read_results("../results/p2/part5pr0.1.csv"), errorbar="sd", ax=axes[1, 1])

plt.savefig("../images/p2/part5.png", bbox_inches="tight")
```

```python
fig, axes = plt.subplots(1, 3, figsize=(10, 3))
probs = [3, 6, 10]
for i in range(3):
    axes[i].set_ylim(65800, 66200)
    axes[i].set_title(f"Population size: {probs[i]}")
sns.lineplot(read_results("../results/p2/part5pop3.csv"), errorbar="sd", ax=axes[0])
sns.lineplot(read_results("../results/p2/part5pop6.csv"), errorbar="sd", ax=axes[1])
sns.lineplot(read_results("../results/p2/part5pop10.csv"), errorbar="sd", ax=axes[2])

plt.savefig("../images/p2/part6.png", bbox_inches="tight")
```
